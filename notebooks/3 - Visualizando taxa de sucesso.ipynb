```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../Dados/taxa-de-sucesso-por-curso-ufc-detalhado.csv', encoding='utf-8', na_values='-')
```

```python
df.columns
```

```python
df['NOME'] = df[['CURSO', 'UNIDADE ACADEMICA', u'PERÍODO']].apply(lambda x: ' | '.join(x), axis=1)
```

```python
hdf = df[['NOME', '2007', '2008', '2009', '2010', '2011', '2012', '2013']].set_index('NOME').sort_index()
```

```python
fig = plt.figure(figsize=(20, 20))
sns.heatmap(hdf, annot=True, fmt=".1f", linewidths=.5)
fig.savefig('taxa-sucesso-heatmap.png', dpi=300)
```

# Por unidade acadêmica

```python
df = pd.read_csv('../Dados/taxa-de-sucesso-por-unidade-academica.csv', encoding='utf-8', na_values='-')
```

```python
df.drop(u'PERÍODO', axis=1, inplace=True)
```

```python
df.set_index(u'UNIDADE ACADÊMICA', inplace=True)
```

```python
fig = plt.figure(figsize=(20, 20))
sns.heatmap(df, annot=True, fmt=".1f", linewidths=.5)
fig.savefig('taxa-sucesso-heatmap.png', dpi=300)
```
