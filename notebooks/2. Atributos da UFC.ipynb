```python
import pandas as pd
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
pd.set_option('display.max_colwidth', 0)
```

```python
def to_pct(x):
    
    if ~np.isnan(x):
        
        return '%.2f%%' % x
    return '-'
```

# Quantitativo de bolsas da graduação

```python
df = pd.read_csv('../Dados/numeros-bolsas-graduacao.csv', na_values='-', encoding='utf-8', index_col=0)
```

```python
total_bolsas = df.sum(axis=1)
```

```python
total_bolsas
```

```python
fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(111)
ax.set_title(u'Total de bolsas para graduação')

ax.plot(total_bolsas.index, total_bolsas, linewidth=3.)

ax.grid()

ax.set_xticklabels(total_bolsas.index)

ax.set_ylim([0, 5000])

ax.figure.savefig(u"../Processo/CRISP-DM/1 - Business Understanding/img/numero_bolsas_graduacao_ufc.png", dpi=300)
```

```python
print df.T.to_latex(na_rep="-")
```

# Orçamento da UFC

```python
df = pd.read_csv('../Dados/orcamento-ufc.csv', encoding='utf-8', na_values='-', index_col=0)
```

```python
fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(111)
ax.set_title(u'Orçamento da UFC - em milhões')

ax.plot(df.index, df[u'Previsão'], linewidth=7., alpha=.6)
ax.plot(df.index, df[u'Executado'], linewidth=1., color='red')

ax.legend([u'Previsão', u'Executado'], loc='best')

ax.grid()

ax.set_xticklabels(df.index)

ax.set_ylim([0, 1200])

ax.figure.savefig(u"../Processo/CRISP-DM/1 - Business Understanding/img/orcamento_ufc.png", dpi=300)
```

```python
print df.dropna().to_latex()
```

# Área construída da UFC

```python
df = pd.read_csv('../Dados/area-construida-ufc.csv', encoding='utf-8', na_values='-', index_col=0)
```

```python
df = df.ix[2008:]
```

```python
fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(111)
ax.set_title(u'Área construída da UFC - mil m²')

ax.plot(df.index, df[u'Área Construída (mil m2)'], linewidth=3.)

ax.grid()

ax.set_xticks(df.index)
ax.set_xticklabels(df.index)

ax.set_ylim([0, 360])

ax.figure.savefig(u"../Processo/CRISP-DM/1 - Business Understanding/img/area_construida_ufc.png", dpi=300)
```
