```python
import pandas as pd
import numpy as np
%matplotlib inline
pd.set_option('display.max_colwidth', 0)
```

```python
def to_pct(x):
    
    if ~np.isnan(x):
        
        return '%.2f%%' % x
    return '-'
```

# Número de assistência estudantil

```python
df = pd.read_csv('../Dados/numeros-assistencia-ao-estudante.csv', na_values='-', encoding='utf-8', index_col=0)
```

```python
df.columns.name = 'Ano'
```

```python
print df.dropna(axis=1).to_latex(na_rep='-')
```

```python
print df[[u'ESPECIFICAÇÃO', '2008', '2009', '2010', '2011', '2012', '2013']].to_latex(na_rep='-')
```

### evolução

```python
df.set_index(u'ESPECIFICAÇÃO').T.plot(figsize=(14, 8))
```

# Orçamento da UFC

```python
df = pd.read_csv('../Dados/orcamento-ufc.csv', encoding='utf-8', na_values='-', index_col=0)
```

```python
df.columns.name = 'Ano'
```

```python
df = df.T
```

```python
df.insert(2,'Evolução do executado', df['Executado'].pct_change().apply(to_pct))
```
