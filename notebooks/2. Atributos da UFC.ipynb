```python
import pandas as pd
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
pd.set_option('display.max_colwidth', 0)
```

```python
def to_pct(x):
    
    if ~np.isnan(x):
        
        return '%.2f%%' % x
    return '-'
```

# Quantitativo de bolsas da graduação

```python
df = pd.read_csv('../Dados/numeros-bolsas-graduacao.csv', na_values='-', encoding='utf-8', index_col=0)
```

```python
total_bolsas = df.sum(axis=1)
```

```python
total_bolsas
```

```python
fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(111)
ax.set_title(u'Total de bolsas para graduação')

ax.plot(total_bolsas.index, total_bolsas, linewidth=3.)

ax.grid()

ax.set_xticklabels(total_bolsas.index)

ax.set_ylim([0, 5000])

ax.figure.savefig(u"../Processo/CRISP-DM/1 - Business Understanding/img/numero_bolsas_graduacao_ufc.png", dpi=300)
```

```python
print df.T.to_latex(na_rep="-")
```

# Orçamento da UFC

# Números de inscritos, vagas ofertadas e demanda

```python
df = pd.read_csv('../Dados/numeros-inscritos-vagas-ofertadas-demanda-ufc.csv', encoding='utf-8', na_values='-', index_col=0)
```

```python
df
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(5, 8))

ax1.set_title(u'Quantidade de candidatos inscritos')

ax1.plot(df.index, df[u'Candidatos Inscritos'], linewidth=3.)

ax1.grid()

ax1.set_xticks(df.index)
ax1.set_xticklabels(df.index)

ax1.set_ylim([0, 140000])

ax2.set_title(u'Quantidade de vagas ofertadas')

ax2.plot(df.index, df[u'Vagas Ofertadas'], linewidth=3.)

ax2.grid()

ax2.set_xticks(df.index)
ax2.set_xticklabels(df.index)

ax2.set_ylim([0, 7000])

ax3.set_title(u'Concorrência')

ax3.plot(df.index, df[u'Demanda'], linewidth=3.)

ax3.grid()

ax3.set_xticks(df.index)
ax3.set_xticklabels(df.index)

ax3.set_ylim([0, 22])

fig.savefig(u"../Processo/CRISP-DM/1 - Business Understanding/img/inscritos_ufc.png", dpi=300)
```

```python
print df.to_latex()
```
